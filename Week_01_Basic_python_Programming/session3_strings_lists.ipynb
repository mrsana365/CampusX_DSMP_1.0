## 2. Loop Control Statements

These statements are used to alter the flow of a loop based on specific conditions.

* **`break`:** Terminates the loop immediately when a condition is met (e.g., exiting when `i == 5`).
* **Prime Number Range:** A practical application of `break`. It uses a nested loop to check for factors; if a factor is found, it "breaks" the inner loop. The `else` block attached to the `for` loop prints the number only if the loop finished without breaking.
* **`continue`:** Skips the current iteration and moves to the next one. Used in the notebook to skip printing the number 5 in a sequence.
* **`pass`:** A null operation used as a placeholder. It tells the compiler to do nothing and move on without throwing an error.


```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

###  Find out the prime number between range

## 3. Creating and Accessing Strings

* **Flexibility:** If a start index is out of range, Python adjusts it to the closest valid position rather than throwing an error.

```python
# 4 ) Write a program which can remove a particular character from a string.
s = input('enter the string')
term = input('what would like to remove')

result = ''

for i in s:
  if i != term:
    result = result + i

print(result)
```

```python
# 5) Write a program that can check whether a given string is palindrome or not.
# abba
# malayalam

s = input('enter the string')
flag = True
for i in range(0,len(s)//2):
  if s[i] != s[len(s) - i -1]:
    flag = False
    print('Not a Palindrome')
    break

if flag:
  print('Palindrome')

```

```python
# 6 )  Write a program to count the number of words in a string without split()

s = input('enter the string')
L = []
temp = ''
for i in s:

  if i != ' ':
    temp = temp + i
  else:
    L.append(temp)
    temp = ''

L.append(temp)
print(L)

```

```python
# 7 )  Write a python program to convert a string to title case without using the title()
s = input('enter the string')

L = []
for i in s.split():
  L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```

```python
# 8 )  Write a program that can convert an integer to string.

number = int(input('enter the number'))

digits = '0123456789'
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10

print(result)
print(type(result))
```
