```python
### fr1.__add__(fr2) == fr1 + fr2
### The Reality: When you write fr1 + fr2 in your main code, Python secretly converts it into this function call: fr1.__add__(fr2)

class Fraction:

  # parameterized constructor
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)
  
fr1 = Fraction( 3,4)
fr2 = Fraction( 1,2)
print ( fr1)        ## 3/4
print ( fr2)        ## 1/2

print( fr1 + fr2)   ##10/8  as TEXT not as the object instance
```

```python
class Fraction:

    # 1. The Constructor (The Blueprint)
    def __init__(self, x, y):
        self.num = x
        self.den = y

    # 2. The Translator (The Interface)
    def __str__(self):
        return '{}/{}'.format(self.num, self.den)

    # 3. The Operator (The Logic)
    def __add__(self, other):
        # Step A: Do the math
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        
        # Step B: CRITICAL CHANGE - Create a NEW object
        # We do NOT return text. We return a new instance of the class.
        return Fraction (new_num, new_den)

# Execution
fr1 = Fraction(3, 4)
fr2 = Fraction(1, 2)

# Now, let's add them
result = fr1 + fr2  
print(result)
```

```python
class Fraction:

  # parameterized constructor
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __sub__(self,other):
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __mul__(self,other):
    new_num = self.num*other.num
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    return '{}/{}'.format(new_num,new_den)

  def convert_to_decimal(self):
    return self.num/self.den




```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

```python
fr1.convert_to_decimal()
# 3/4
```
