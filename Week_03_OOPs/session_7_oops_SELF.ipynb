```python
## Then the big Question is who is "self" ? 
class Robot:
 
    def __init__ (self):
        print ( "id of self--> ", id (self))  # id of self-->  2039610022208

r1=Robot()

print ( "obj r1  id --> " , id (r1)) # obj r1  id -->  2039610022208

# so the "self" is current object as same as obj r1  , bcoz both having same id 
# so if one method want to called another method so first it called to 
# constructor -->> then constructor called to object  == self as agrumemt to other function


# ---------------  OUTPUT  -------------------------------------

# id of self-->  2039610022208
# obj r1  id -->  2039610022208


```

```python

class Robot:
 
    def __init__ (self):
        print ( "id of self--> ", id (self))   # id of self-->  2039610022208

    def speak():
        print ( "I am speak method that give feature to robot")


r1 = Robot()  # obj r1  id -->  2039610022208

r1.speak()


# ---------------  OUTPUT  -------------------------------------------
# id of self-->  2039610021536

# TypeError: Robot.speak() takes 0 positional arguments but 1 was given


##------------ Conclusion ------------------------------------------------------------------------------------
## we can conclude that when we run this  "r1.speak()"  r1 is pass to the speak() method as argument in behind 



 
```

```python
class Robot:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"Hi! My name is {self.name}."

r1 = Robot("Bender")   

# Python calls the __init__ method to set the name
Robot.__init__(r1, "Bender")

print(r1.speak())   ## r1.speak() ===  Robot.speak(r1) 
Robot.speak(r1) 
# Output: Hi! My name is Bender.


```
